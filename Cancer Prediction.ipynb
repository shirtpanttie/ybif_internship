```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Cancer.csv')
```

```python
df.head()
```

```python
x=df.drop(['id', 'diagnosis','Unnamed: 32'],axis=1)
```

```python
x.shape
```

```python
x
```

```python
y = df['diagnosis']
```

```python
y
```

```python
y.shape
```

```python
xtr,xte,ytr,yte = train_test_split(x,y,test_size=0.2,random_state=2529)
```

```python
xtr,xte,ytr,yte
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(xtr,ytr)
```

```python
ypred = model.predict(xte)
```

```python
ypred
```

```python
accuracy_score(yte,ypred)
```

```python
confusion_matrix(yte,ypred)
```

```python
print(classification_report(yte,ypred))
```
