```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
y = df['diabetes']
x = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
```

```python
x_train,x_test,y_train,y_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
```

```python
model.fit(x_train,y_train)
# the ouptu is a warning not an error to overcome include max_iter
# in the above step
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
# accuracy score not mean error or mean square error
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
